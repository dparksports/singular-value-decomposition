```python
# compute F from x1, x2
# rank2 F
# normalize F

A = np.zeros((8,9))
for i in range(8):
    rowA = np.array([[x1[i,0]*x2[i,0], x1[i,0]*x2[i,1], x1[i,0]],
    [x1[i,1]*x2[i,0], x1[i,1]*x2[i,1], x1[i,1]],[x2[i,0], x2[i,1], 1]])
#     print(rowA)
    A[i,:] = np.reshape(rowA, (9,))

print(A, A.shape,"A")
```

```python
U,S,V = np.linalg.svd(A, full_matrices=True)
print(V, V.shape, "V")
```

```python
Varray = V[:,-1]
print(Varray, Varray.shape)
```

```python
F = V[:,-1].reshape((3,3))
print(F)
```

```python
U_hat,S_hat,V_hat = np.linalg.svd(F, full_matrices=True)
# print(S_hat.shape,S_hat)
S_hat[-1] = 0 
D = np.diag(S_hat)
F_rank2 = np.dot(U_hat, np.dot(np.diag(S_hat), V_hat.T))
F = F_rank2
print(F, F.shape, "F")
```

```python
F_norm = np.linalg.norm(F)
print(F_norm)
F_normalized = F / np.linalg.norm(F)
F = F_normalized
print(F_normalized, F_normalized.shape, "F_normalized")
```

```python
# compute E from F, K
# normalize E

print(K)
print(K.T)

E_noisy = K.T.dot(F.dot(K))
print(E_noisy)

U,singular_values_noisy,V = np.linalg.svd(E_noisy, full_matrices=True)
print(singular_values_noisy.shape,singular_values_noisy)

singular_values_noisy[0] = singular_values_noisy[1] = 1 
singular_values_noisy[-1] = 0 

singular_values_corrected = np.diag(singular_values_noisy)
print(singular_values_corrected.shape,singular_values_corrected)

E_corrected = U.dot(singular_values_corrected.dot(V.T))
E = E_corrected
print(E_corrected)

E_norm = np.linalg.norm(E)
print(E_norm)
E_normalized = E / np.linalg.norm(E)
E = E_normalized
print(E_normalized)
```
