```python
pa = np.array([0,0,5,-5])
pa = pa.reshape(2,-1)
print(pa,pa.shape,'pa')

pb= np.array([0,0,5,1])
pb= pb.reshape(2,-1)
print(pb,pb.shape,'pb')

pc = np.array([0,0,-3,3])
pc = pc.reshape(2,-1)
print(pc,pc.shape,'pc')
```

```python
a = pa[1,:]
a = a.reshape(-1)
print(a,a.shape,'a')

b = pb[1,:]
b = b.reshape(-1)
print(b,b.shape,'b')

c = pc[1,:]
c = c.reshape(-1)
print(c,c.shape,'c')
```

```python
import matplotlib.pyplot as plt

plt.plot(pa[:,0],pa[:,1],label='a')
plt.plot(pb[:,0],pb[:,1],label='b')
plt.plot(pc[:,0],pc[:,1],label='c')
plt.grid()
```

```python
dot = np.dot(b,c) 
arccos = np.dot(b,c) / (np.linalg.norm(b) * np.linalg.norm(c))
theta = np.arccos(arccos)
# print(dot)
print(theta)
print(np.cos(theta))
```
