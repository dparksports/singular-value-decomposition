```python
import numpy as np
```

```python
a = np.array([0,1,2.2,2.9])
a = a.reshape(-1,1)
a
```

```python
b = np.array([-0.8,0,0.9,2.1])
b = b.reshape(-1,1)
b
```

```python
x = np.arange(0,3,1)
x
```

```python
x1 = np.array([0.59/0.58, 0])
y1 = np.array([0, 0.58/-0.57])
y1
```

```python
x2 = np.array([0.57/-0.59, 0])
y2 = np.array([0, 0.57/0.58])
y2
```

```python
x3 = np.array([1, 0])
y3 = np.array([0, -1])
y3
```

```python
print(a,a.shape,'a')
print(b,b.shape,'b')
```

```python
x4 = a
y4 = b
print(x4,'x4')
print(y4,'y4')
```

```python
U,S,VT = np.linalg.svd(a,full_matrices=False)
print(U,U.shape,"U")
print(S,S.shape,"S")
print(VT,VT.shape,"VT")
```

```python
pseudo = VT.T @ np.linalg.inv(np.diag(S)) @ U.T @ b
print(pseudo,pseudo.shape,"pseudo")
```

```python
import matplotlib.pyplot as plt
plt.plot(x1,y1,label='x1')
#plt.plot(x2,y2,label='x2')
#plt.plot(x3,y3,label='x3')

plt.plot(x4,y4,'x',Color='r',label='noisy data')
plt.plot(x,pseudo[0,0]*x,'--',Color='b',label='A pseudo - LS')
plt.grid()
plt.legend()
```
